```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("unemployment in india.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.rename(columns={
    'Region': 'State',
    ' Date ': 'Date',
    'Estimated Unemployment Rate (%)': 'Unemployment_Rate',
    'Estimated Employed': 'Employed',
    'Estimated Labour Participation Rate (%)': 'Labour_Participation_Rate'
}, inplace=True)
```

```python
print(df.columns)
```

```python
df.columns = df.columns.str.strip()
print(df.columns)
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
```

```python
print(df.columns)
```

```python
statewise_unemployment = df.groupby("state")["estimated_unemployment_rate_(%)"].mean().sort_values(ascending=False)
```

```python
plt.figure(figsize=(6, 5))
sns.barplot(x=statewise_unemployment.values, y=statewise_unemployment.index, palette="coolwarm")
plt.title("Average Unemployment Rate by State")
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(5, 6))

sns.lineplot(
    data=df,
    x='date',
    y='estimated_unemployment_rate_(%)',
    hue='state',
    legend=False
)

plt.title("Unemployment Rate Over Time (by State)")
plt.xlabel("Date")
plt.ylabel("Estimated Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
selected_states = ['Tamil Nadu', 'Maharashtra', 'Delhi', 'Kerala']
filtered_df = df[df['state'].isin(selected_states)]

sns.lineplot(
    data=filtered_df,
    x='date',
    y='estimated_unemployment_rate_(%)',
    hue='state'
)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
heatmap_data = df[[
    'estimated_unemployment_rate_(%)',
    'estimated_employed',
    'estimated_labour_participation_rate_(%)'
]]
corr = heatmap_data.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
area_unemployment = df.groupby('area')['estimated_unemployment_rate_(%)'].mean()
fig, ax = plt.subplots()
ax.pie(area_unemployment, 
       labels=area_unemployment.index, 
       autopct='%1.1f%%', 
       startangle=90)
ax.set_title("Unemployment Rate by Area Type")
plt.axis('equal')  
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.lineplot(data=df, x='date', y='estimated_unemployment_rate_(%)', hue='state')
plt.title("Unemployment Rate Over Time by State")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5, 5))
sns.histplot(df['estimated_unemployment_rate_(%)'], bins=20, kde=True, color='skyblue')
plt.title("Distribution of Unemployment Rate")
plt.xlabel("Unemployment Rate (%)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
